```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
sc_y=StandardScaler()
#In standardscaler fit method will calculate the mean and s_d and store it 
#in object sc_x sc_y and transform method is used to standardize the data 
#using the data(mean and sd ) obtanined from fit method;
x=sc_x.fit_transform(x)
y=sc_y.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(x,y,random_state=2022,test_size=0.2)
```

```python
from sklearn.svm import SVR
regressor=SVR()#by default kernel is rbf hence higher score
regressor.fit(x_train,y_train)
y_pred=regressor.predict(x_test)
regressor.score(x_test,y_test)
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='linear')
regressor.fit(x_train,y_train)
y_pred=regressor.predict(x_test)
regressor.score(x_test,y_test)
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='poly')
regressor.fit(x_train,y_train)
y_pred=regressor.predict(x_test)
regressor.score(x_test,y_test)
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='rbf')
regressor.fit(x_train,y_train)
y_pred=regressor.predict(x_test)
regressor.score(x_test,y_test)
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='rbf',C=12,gamma=0.7)
regressor.fit(x_train,y_train)
y_pred=regressor.predict(x_test)
regressor.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor.fit(x_train,y_train)
ypred= regressor.predict(x_test)
print(regressor.score(x,y))
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=100,random_state=2022)
regressor.fit(x_train,y_train)
print("Score is ",regressor.score(x,y))
```
