```python
import os
import csv
import pickle
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor

rear=[]


# Set the path of the folder containing input CSV files
input_folder_path = "/content/drive/MyDrive/github/collegeproject/datasets/unique_crops_south"

# Set the path of the folder to save output pickle files
output_folder_path = "/content/drive/MyDrive/github/collegeproject/pickle_files/crop_yield_south"

# Set the path of the folder to save scaler instances
scaler_folder_path = "/content/drive/MyDrive/github/collegeproject/pickle_files/scalers_crop_yield"

# Loop through all files in the input folder
for filename in os.listdir(input_folder_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(input_folder_path, filename)

        # Load the CSV file as a pandas dataframe
        df = pd.read_csv(file_path)
        # df.drop(['State_Name','PH', 'District_Name','crop', 'Crop_Year'], axis=1, inplace=True)
        # # Convert categorical variable 'Season' to numerical using one-hot encoding
        # df = pd.concat([df, pd.get_dummies(df['Season'])], axis=1)
        # df.drop(['Season'], axis=1, inplace=True)
        # # Prepare the input features and output labels
        # X = df.drop(['Production'], axis=1)
        X=df.iloc[:,[4,6,8]]
        y = df['Production'].values.reshape(-1, 1)
        
        # Scale the data
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)

        # Save the scaler instance as a pickle file
        scaler_file_path = os.path.join(scaler_folder_path, f"{filename}_scaler.pkl")
        with open(scaler_file_path, "wb") as f:
            pickle.dump(scaler, f)
        
        # Split the data into training and testing sets
        x_train, x_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=2022)

        # Train a Decision Tree Regressor
        regressor=RandomForestRegressor(n_estimators=100,random_state=2022)
        regressor.fit(x_train, y_train)
        
        # Evaluate the model on the test data
        score = regressor.score(x_test, y_test)
        rear.append(score)
        print(f"Filename: {filename}, Score: {score}")
        
        # Save the trained model as a pickle file
        output_file_path = os.path.join(output_folder_path, f"{filename}.pkl")
        with open(output_file_path, "wb") as f:
            pickle.dump(regressor, f)
```

```python

def average(lst):
    total = sum(lst)
    return total / len(lst)
avg = average(rear)
print(avg) 
```
