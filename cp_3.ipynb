```python
crops=pd.read_csv("/content/drive/MyDrive/github/collegeproject/datasets/crops_yield_kar.csv")
```

```python
# Get the unique values of the column you want to split the data by
column_values = crops['crop'].unique()
# Loop over the unique values of the column
for value in column_values:
    # Create a dataframe for each unique value
    value_df = crops[crops['crop'] == value]
    # Write the dataframe to a new csv file
    value_df.to_csv(f"{'crop'}_{value}.csv", index=False)
```
